```python
# Pymongo web scraping starts here
import pymongo
import os
import numpy as np
from bs4 import BeautifulSoup as bs
```

```python
# Read HTML from file
filepath = os.path.join("Resources", "template.html")
with open(filepath) as file:
    html = file.read()
```

```python
soup = bs(html, 'lxml')
type(soup)
```

```python
# Print formatted version of the soup
print(soup.prettify())
```

```python
# Extract the title of the HTML document
soup.title
```

```python
# Extract all historical share prices
extract_prices = soup.tbody.find_all('td')
extract_prices
```

```python
pd.DataFrame.from_dict(extract_prices, dtype=None, columns=None)
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the 'classDB' database in Mongo
db = client.sharesyPrice
```

```python
# Query all students
# Here, db.students refers to the collection 'classroom '
sharePrice = db.classroom.find()

# Iterate through each student in the collection
for share in sharePrice:
    print(share)
```
